```python
import tweepy
import pandas as pd

#insert api key and token access
consumer_key = 'EsKaq7Xq3HwERrascXyTJhbtn'
consumer_secret = '5cjl72qg0tZOOoCzwWxa7YAQYi4FRPheZdcEN0ttw3BFrVfFo4'
access_token = '1097888766256046080-HOGzU3XEW9mjHF6Z87imNPkBrIRLSI'
access_token_secret = 'ncxOUkBHjaRhEmz2Nfp4YjtEpJPlf8RluPrf1KCmeInQR'

#tweepy authorization/calling the api 
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth, wait_on_rate_limit=True)

def crawling(keyword):
  #specific tweet
  search_words = keyword
  # new_search = search_words + " -filter:retweets" 
  # date_since = "2021-02-13" #optional: pull tweets since the date xx
  # date_until = "2021-02-22" #optional: pull tweets until the date xx 

  tweets = tweepy.Cursor(api.search,
                          q=search_words,
                          # since=date_since,
                          # until = date_until,
                          geo_enabled='True',     
                          # geocode = coordinates,  
                          tweet_mode='extended', 
                          lang='en'               
                          ).items(1000)             

  usid = []
  date = []
  username = []
  twid = []
  folcount = []
  frcount = []
  items = []
  mentions = []
  retweet = []
  favcount = []

  for tweet in tweets:
    if hasattr(tweet, "entities"):
      entities = tweet.entities
      if "user_mentions" in entities:
        for ent in entities["user_mentions"]:
          if ent is not None:
            if "screen_name" in ent:
              screen_name = ent["screen_name"]
              if screen_name is not None:
                usid.append(str(tweet.id_str))
                date.append(tweet.created_at)
                mentions.append(screen_name)
                username.append(tweet.user.screen_name)
                items.append(tweet.full_text.encode('utf-8'))
                twid.append(str(tweet.id))
                folcount.append(tweet.user.followers_count)
                frcount.append(tweet.user.friends_count)
                retweet.append(tweet.retweet_count)
                favcount.append(tweet.favorite_count)
  
  hasil = pd.DataFrame(list(zip(usid, date, username, items, twid, mentions, folcount, frcount, retweet, favcount)),
                    columns=['User ID', 'Date', 'Username From', 'Tweet','Tweet ID','Mention', 'followers_count', 'friends_count', 'retweet_count', 'favorite_count'])

  return hasil
```
